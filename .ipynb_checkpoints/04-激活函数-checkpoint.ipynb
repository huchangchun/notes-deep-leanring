# 激活函数的作用

激活函数不是真的去激活什么。在神经网络中，激活函数的作用是能够给神经网络以更好的解决较为复杂的问题。线性模型的表达能力不够，引入激活函数是为了添加非线性因素。
比如在下面这个问题中：
![0401](./data/04-01.png)
如上图，在最简单的情况下，数据是线性可分的，只需要一条直线就已经可以对数据进行分类
![0402](./data/04-02.png)
但是情况变得复杂一点呢？数据就变成了线性不可分的情况显然直线不能对样本进行很好的分类了。
![0403](./data/04-03.png)
于是我们尝试引入非线性的因素，对样本进行分类。
在神经网络中也类似，我们需要引入一些非线性的因素，来更好的解决复杂的问题，而激活函数，它使得我们的神经网络能更好的解决较为复杂的问题。

### 激活函数的定义及其相关概念

什么是饱和？说人话就是：指当输入比较大的时候几乎没有变化了，硬饱和就是单位跃阶函数输入非0输出直接干到极限值，软饱和指输入趋于无穷时才接近极限值，饱和就是导致梯度消失）
- 饱和： 当一个激活函数$h(x)$满足$$\lim_{n \rightarrow + \infty} h^{'}(x)=0$$
时我们称之为右饱和。
当一个激活函数$h(x)$满足$$\lim_{n \rightarrow - \infty} h^{'}(x)=0$$
时我们称之为左饱和。当一个激活函数既满足左饱和又满足右饱和时，我们称之为饱和。
- 硬饱和与软饱和：
对于任意的$x$，如果存在常熟$c$，当$x>c$时恒有$h^{'}(x)=0$则称其为右硬饱和，并且$x<c$时满足左硬饱和，则我们称之为硬饱和.
一个激活函数在极限处梯度才达到0时称为软饱和。（. (Hard and Soft Saturation). Let c be a
constant such that x > c implies h0
(x) = 0 and left hard
saturates when x < c implies h0
(x) = 0, ∀x. We say that
h(·) hard saturates (without qualification) if it both left and
right hard saturates. An activation function that saturates but
achieves zero gradient only in the limit is said to soft saturate.）
引自论文：https://arxiv.org/pdf/1603.00391v3.pdf


### Sigmoid函数
Sigmoid函数曾被广泛第应用，但由于其自身的一些缺陷，现在很少被使用了，Sigmoid函数$f(x) = \frac{1}{1+e^{-x}}$
对应的图像如下：

```python
%matplotlib inline
import numpy as np
def sigmoid(x):
    return 1/(1+np.exp(-x))
import matplotlib.pyplot as plt
x = np.arange(-10,10,0.1)
y = [sigmoid(i) for i in x]
plt.figure()
plt.plot(x,y)
plt.grid(True)
plt.show()
```

```python
plt.close()
```

#### Sigmoid函数优点：
- Sigmoid函数的输出映射在（0,1）之间，单调连续，输出范围有限，优化稳定，可以用作激活函数。
- 求导容易。

缺点：
- 由于其软饱和性（输入趋于无穷大的时候输出无限接近上限），容易产生梯度消失，导致训练出现问题。
- 其输出并不是以0为中心的。

### tanh函数
现在比起Sigmoid函数我们通常更倾向于tanh函数。tanh函数被定义为$$f(x) = \frac{1-e^{-2x}}{1+e^{-2x}}$$
其图像为位于[-1,1]区间上：

```python
%matplotlib inline
import numpy as np
def sigmoid(x):
    return (1-np.exp(-2*x))/(1+np.exp(-2*x))
import matplotlib.pyplot as plt
x = np.arange(-10,10,0.1)
y = [sigmoid(i) for i in x]
plt.figure()
plt.plot(x,y)
plt.grid(True)
plt.show()
```

plt.close()

#### tanh函数优点：
- 比Sigmoid函数收敛速度更快
- 相比Sigmoid函数其输出以0为中心。

缺点：
- 还是没有改变Sigmoid函数的最大问题--由于饱和性产生的梯度消失问题。

### ReLU函数
ReLU是最近几年非常受欢迎的激活函数，被定义为 $y=\begin{cases}0 \space(x\leq0)\\x \space(x\gt 0)\end{cases}$

```python
%matplotlib inline
import numpy as np
def sigmoid(x):
    return 0 if x <= 0 else x 
import matplotlib.pyplot as plt
x = np.arange(-10,10,0.1)
y = [sigmoid(i) for i in x]
plt.figure()
plt.plot(x,y)
plt.grid(True)
plt.show()
```

```python
plt.close()
```

#### ReLU函数优点：
- 1.相比Sigmoid和tanh，ReLU更快地到达线性、非饱和的形式。
- 2. 有效缓解了梯度消失的问题。
- 3.在没有无监督预训练的时候也能较好的表现。
- 4.提供了神经网络的稀疏表达能力。
- 5.ReLU简单实用。
缺点：
随着训练的进行，可能会出现神经元死亡，权重无法更新的情况。

#### 目前卷积神经网络RELU有多种形式的变种并用在卷积层，ELU,LReLU,RELU,SReLU,这里不再做扩展。 
CNN激励层建议：
- CNN尽量不要实用Sigmod，如果要使用，建议只在全连接层使用。
- 首先使用RELU,因为迭代速度快，但是可能效果不佳。
- 如果使用RELU实效的情况下，考虑使用Leaky ReLU或者Maxout，此时一般情况都可以解决。
- tanh激活函数在某些情况下有比较好的效果，但是应用场景较少。

关于激活函数的选择，目前没有标准答案，需要结合实际情况
